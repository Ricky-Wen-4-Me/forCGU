```python
import matplotlib.pyplot as pl
import numpy as np

X = np.random.normal(size = 1000)

pl.figure()

dx = 0.1
xrange = np.arange(-3, 3, dx)

pl.stem(X)
pl.xlabel('x')
pl.ylabel('f(x)')
pl.grid()
```

```python
dx = 0.1
xrange = np.arange(-3, 3, dx)

pl.hist(X)
```

```python
from sympy.stats import variance 

means = np.mean(X)
variance = np.var(X)

means, variance
```

```python
X_sort = np.sort(X)
pl.figure()

dx = 0.1
xrange = np.arange(-3, 3, dx)

ff = []

for x in xrange:
    q = X_sort[(X_sort >= x)&(X_sort < x+dx)].size
    ff += [(x,q)]
    
ff = np.array(ff)

xx = ff[:,0]
yy = ff[:,1]

pl.stem(xx, yy)
pl.xlabel('x')
pl.ylabel('f(x)')
pl.grid()
```
